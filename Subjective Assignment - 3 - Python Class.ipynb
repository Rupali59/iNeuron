Q3. Write the Python class to find the validity of the string of the
parentheses, '(', ')', '{', '}', '[' and '] and the brackets must be closed
in the correct order, example - "()" and "()[]{}" are valid but "[)",
"({[)]" and "{{{" are invalid.


```python
def check(my_string): 
    brackets = ['()', '{}', '[]'] 
    while any(x in my_string for x in brackets): 
        for br in brackets: 
            my_string = my_string.replace(br, '') 
    return not my_string 
   
# string = "{[]{()}}"
# string = "({[)]"
string = "{{{"
print(string, "-", "Valid" if check(string) else "Invalid") 
```

Q4. Write the Python class to get all possible unique subsets from a set
of distinct integers?

    Input - [4, 5, 6]
    Output - [[], [6], [5], [5, 6], [4], [4, 6], [4, 5], [4, 5, 6]]

```python
import itertools 

def findsubsets(s): 
    subsets= [[]]
    for j in range(1,len(s)+1):
        for i in itertools.combinations(s, j):
            subsets.append(list(set(i)))
    return subsets

words = [int(x) for x in input("Input: ").split(", ")]
print("Output - "+list(findsubsets(words)))
```

Q5. Write the Python class to find a pair of elements (indices of the two
numbers) from a given array whose sum equals the specific target
number?

    Input: numbers- [10,20,10,40,50,60,70], target=50
    Output- 3, 4

```python
def target_index(numbers, target):
    for i in range(0, len(numbers)):
        for j in range(i+1, len(numbers)):
            if(numbers[i]+numbers[j] == target):
                return [i,j]
    return None

numbers = [int(x) for x in input("numbers- ").split(",")]
target = int(input("target= "))
print("Output - "+', '.join(map(str, target_index(numbers, target))))
```

Q6. Write the Python class to find the three elements that sum to zero
from the set of n real numbers?

    Input array- [-25, -10, -7, -3, 2, 4, 8, 10]
    Output - [[-10, 2, 8], [-7, -3, 10]]


```python
numbers = [-25, -10, -7, -3, 2, 4, 8, 10]#[int(x) for x in input("Input array- ").split(",")]
output=[]
for i in range(0, len(numbers)):
    for j in range(i+1, len(numbers)):
        for k in range(j+1, len(numbers)):
            if(numbers[i]+numbers[j]+numbers[k] == 0):
                output.append([numbers[i],numbers[j],numbers[k]])
print("Output - "+', '.join(map(str, output)))
```

Q7. Write the Python class to implement pow(x, n)?

```python
def pow(x, n):
    p = 1
    if n < 0:
        x = 1 / x
        n = abs(n)

    # Exponentiation by Squaring
    while n:
        if n % 2:
            p *= x
        x *= x
        n //=2
    return p

print(pow(2, 2))
print(pow(2, -2))
print(pow(-2, 2))
print(pow(-2, -2))
```

Q10. Write the Python class named Circle constructed by the radius
and two methods which will compute the area and perimeter of
the circle?

```python
import math

class Circle:
    radius=0
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi*self.radius*self.radius
        
    def perimeter(self):
        return 2*math.pi*self.radius

circle = Circle(7)
print("Perimeter: "+str(circle.perimeter()))
print("Area: "+str(circle.area()))
```

Q12. Write the Python program to count the number of students of
individual class?

    Sample dataclasses = (
    ('V', 1),
    ('VI', 1),
    ('V', 2),
    ('VI', 2),
    ('VI', 3),
    ('VII', 1),
    )
    Expected OutputCounter({'VI': 3, 'V': 2, 'VII': 1})

```python
import collections
dataclasses = (
('V', 1),
('VI', 1),
('V', 2),
('VI', 2),
('VI', 3),
('VII', 1),
)

dct = collections.defaultdict(int)
for _class, students in dataclasses:
        if dct[_class]:
            if students>dct[_class]:
                dct[_class] = students
        else:
            dct[_class] = students
        
print("Output: ")
print(dict(dct) )
```
